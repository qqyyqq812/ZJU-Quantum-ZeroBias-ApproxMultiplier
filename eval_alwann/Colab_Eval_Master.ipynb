## Step 4: 异构作图雷达
无需等待上面的训练格结束！您可以 Stop 上面的执行，到这里出图后，再回去继续练。

```python
import json
import matplotlib.pyplot as plt
import os

f_path = 'results_local/eval_history.json'
if os.path.exists(f_path):
    with open(f_path, 'r') as f:
        data = json.load(f)
    
    cos_sims = data['cos_sims']
    plt.figure(figsize=(10,5))
    plt.plot(cos_sims, alpha=0.85, color='teal', linewidth=2)
    plt.title(f"Real-Time ImageNet Proxy Accuracy (Tracking Step: {data['step']})")
    plt.xlabel("Inference Batch")
    plt.ylabel("Cosine Similarity (ALWANN vs Baseline)")
    plt.grid(True, linestyle='--')
    plt.show()
else:
    print("JSON 未就绪，尚未捕获有效雷达快照。")
```
